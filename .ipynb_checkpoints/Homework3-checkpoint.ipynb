```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
import numpy as np
from sklearn.externals.six import StringIO  
import pydotplus
from IPython.display import Image  
import os

from sklearn import metrics

from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
```

```python
data_file_name = './Titanic.csv'
```

Independent Variables: pclass, sex, age, sibsp | Dependent Variables: Survived

```python
data = pd.read_csv (data_file_name)

y = data.survived

data = data.drop('name', axis=1)
data = data.drop('parch', axis=1)
data = data.drop('ticket', axis=1)
data = data.drop('fare', axis=1)
data = data.drop('cabin', axis=1)
data = data.drop('embarked', axis=1)
data = data.drop('boat', axis=1)
data = data.drop('body', axis=1)
data = data.drop('home.dest', axis=1)

data = data.dropna(axis=1)

data.head()
```

```python
le = LabelEncoder()

data['pclass'] = le.fit_transform(data['pclass'])

data['sex'] = le.fit_transform(data['sex'])
```

```python
data_x = data
data_x = data_x.drop('survived', axis=1)

X_train, X_test, y_train, y_test = train_test_split(data_x, y, test_size=0.3)
```

```python
model = tree.DecisionTreeClassifier()

model.fit(X_train, y_train)
model.predict(X_test)

dot_data = StringIO()
tree.export_graphviz(model, out_file=dot_data, filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

```python
score_training = model.score(X_train, y_train)
print('In-sample percent survivors and fatalities correctly predicted:')
print(score_training)

score_test = model.score(X_test, y_test)
print('Out-of-sample percent survivors and fatalities correctly predicted:')
print(score_test)
```

```python
grid = {'max_depth':[3,4,5,6]}
grid_search = GridSearchCV(model, grid, cv=10, refit=True, 
                                               return_train_score=True)
grid_search.fit(data_x, y)
results = grid_search.cv_results_

print(results.get('params'))
print(results.get('mean_test_score'))
print(results.get('mean_train_score'))
print(grid_search.best_params_)
print(grid_search.best_score_)
final_model = grid_search.best_estimator_
```

The tree would be a depth of 3

```python
dot_data = StringIO()
tree.export_graphviz(final_model, out_file=dot_data, filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
