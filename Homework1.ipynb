```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
noise_scale = 100
number_of_samples = 50

x = 25*(np.random.rand(number_of_samples, 1) - 0.8)
y = 5 * x + 20 * x**2 + 1 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
plt.show()
```

```python
x = np.array(x)
y = np.array(y)

x_flattened = x.flatten()
y_flattened = y.flatten()
```

<b>Part 1:</b>
Used an m value of 3

```python
z_3 = np.polyfit(x_flattened,y_flattened,3)
```

```python
z_3_poly = np.poly1d(z_3)
print(z_3_poly)
```

```python
x_pred = np.arange(min(x), max(x))
y_pred_3 = z_3_poly(x_pred)
plt.plot(x_pred, y_pred_3, 'b')
plt.plot(x,y,'ro')
plt.show()
```

<b>Part 2:</b>

```python
x_sorted = np.sort(x_flattened)

z_1 = np.polyfit(x_flattened,y_flattened,1)
z_2 = np.polyfit(x_flattened,y_flattened,2)
z_4 = np.polyfit(x_flattened,y_flattened,4)
z_5 = np.polyfit(x_flattened,y_flattened,5)
z_6 = np.polyfit(x_flattened,y_flattened,6)
z_7 = np.polyfit(x_flattened,y_flattened,7)
z_8 = np.polyfit(x_flattened,y_flattened,8)

y_sorted = np.sort(y_flattened)
```

```python
def findPolynomial(coefficients):
    return np.poly1d(coefficients)

def findYValues(x_values, polynomial):
    return polynomial(x_values)

def findMSE(y_true, y_predicted):
    return np.square(np.subtract(y_true, y_predicted)).mean()

MSE1 = findMSE(y_sorted, findYValues(x_sorted, findPolynomial(z_1)))
MSE2 = findMSE(y_sorted, findYValues(x_sorted, findPolynomial(z_2)))
MSE3 = findMSE(y_sorted, findYValues(x_sorted, findPolynomial(z_3)))
MSE4 = findMSE(y_sorted, findYValues(x_sorted, findPolynomial(z_4)))
MSE5 = findMSE(y_sorted, findYValues(x_sorted, findPolynomial(z_5)))
MSE6 = findMSE(y_sorted, findYValues(x_sorted, findPolynomial(z_6)))
MSE7 = findMSE(y_sorted, findYValues(x_sorted, findPolynomial(z_7)))
MSE8 = findMSE(y_sorted, findYValues(x_sorted, findPolynomial(z_8)))

m = np.arange(1,9)
MSE_Values = [MSE1, MSE2, MSE3, MSE4, MSE5, MSE6, MSE7, MSE8]

plt.plot(m, MSE_Values)
plt.show()
```

The best value of m is 3
